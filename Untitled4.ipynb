```python
import pandas as pd
new_file_path = r"C:\Users\projects\melb_data.csv"
sales_data = pd.read_csv(new_file_path)
```

```python
sales_data.describe()
```

```python
sales_data.columns
```

```python
sales_data= sales_data.dropna(axis=0)
```

```python
y = sales_data.Price
sales_features =['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x= sales_data[sales_features]
```

```python
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
sales_model = DecisionTreeRegressor(random_state=1)
```

```python
sales_model.fit(x,y)
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(sales_model.predict(x.head()))
```
